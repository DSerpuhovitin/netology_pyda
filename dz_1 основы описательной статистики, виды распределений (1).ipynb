Построить частотный график для двух параметров: Freedom и positive affect. Какое это распределение?

Найти медиану, моду, математическое ожидание, стандартное отклонение,  дисперсию, минимум и максимум, найти выбросы (если есть), размах для колонки Freedom

```python
#медиана
print('Медиана: ', statistics.median(df['Freedom']))
```

```python
#мода

freedom = df['Freedom']
freedom_counts = {}
for p in freedom:
    if p not in freedom_counts:
        freedom_counts[p] = 1
    else:
        freedom_counts[p] += 1

# Проходимся по словарю и ищем максимальное количество повторений
# Алгоритм поиска максимума
maxp = 0
mode_freedom = None
for k, v in freedom_counts.items():
    if maxp < v:
        maxp = v
        mode_freedom = k
print('Значение моды: ', mode_freedom, 'Количество встречаемости: ', maxp)
```

```python
#математическое ожидание

print('Способ 1: ',statistics.mean(df['Freedom']))

print('Способ 2: ',sum(df['Freedom'].values)/len(df['Freedom'].values))
```

```python
#стандартное отклонение
freedom = df['Freedom'].values
statistics.pstdev(freedom)
```

```python
#дисперсия
freedom = df['Freedom']
statistics.pvariance(freedom)
```

```python
#минимум и максимум
freedom = df['Freedom']
print('MIN = ', min(df['Freedom']), '| ' 'MAX = ', max(df['Freedom']))
```

```python
#размах
freedom = df['Freedom']
freedom_range = max(freedom) - min(freedom)
print(freedom_range)
```

Посмотреть корреляцию между Freedom и positive affect (графически и математически). Графически - plt.scatter, математически - посчитать коэффициент Пирсона. О чем говорит знак коэффициента?

```python
#корреляция между Freedom и positive affect
freedom = df['Freedom']
positive_affect = df['Positive affect']
plt.xlabel('выборка duration')
plt.ylabel('выборка avg_vote')
plt.title('Точечный график зависимости двух выборок x и y')
plt.grid(True)
plt.scatter(freedom, positive_affect, marker='.')
```

```python
#коэффициент Пирсона
from scipy.stats import pearsonr

freedom = df['Freedom']
positive_affect = df['Positive affect']
corr, p_value = pearsonr(freedom, positive_affect)
plt.scatter(freedom, positive_affect, marker='.')
print('Коэффициент корреляции Пирсона:', corr)
```

Нормализовать данные колонки Freedom (централизация + масштабирование)

```python
freedom = df['Freedom']

print('Центрирование:' , statistics.mean(freedom))

freedom_norm = freedom - statistics.mean(freedom)
print('Нормализация:' , statistics.mean(freedom_norm))

print()
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))
freedom_norm_scale = min_max_scaler.fit_transform(np.array(freedom_norm).reshape(-1, 1))
print(min(freedom_norm_scale))
print(max(freedom_norm_scale))
```
