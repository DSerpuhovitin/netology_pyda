```python
import numpy as np
import pandas as pd
import random

from sklearn.linear_model import LinearRegression
```

```python
# Создаём сэмпл
n_samples = 500

age_owner = np.random.choice(90, n_samples) + 18
length = np.random.choice(120, n_samples) + 15
width = np.random.choice(80, n_samples) + 10

#Собственные признаки
# пллощадь
area = width * length/10

#рассояние до центра и предположим, что оно зависит от возраста владельца (молодёж предпочитает селиться ближе к центру)
dist_center  = np.random.choice(50, n_samples) * age_owner/70

#защищённость в случае зомби-апокалипсиса: чем ближе к центру, тем хуже
zombi_def = (np.random.sample(n_samples) + 0.1) * dist_center

price = 20*area + 15*dist_center + 25*zombi_def + 5*age_owner.mean()

data = pd.DataFrame({'age_owner': age_owner, 'length': length, 'width': width, 'area': area,'dist_center' : dist_center,
                     'zombi_def' : zombi_def, 'price': price})
data.head(10)
```

```python
from sklearn.metrics import mean_absolute_error

X = data[['age_owner', 'area', 'dist_center', 'zombi_def']]
y = data['price']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['age_owner', 'area', 'dist_center', 'zombi_def']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
#Поробуем с другим набором признаков

X = data[['age_owner', 'area', 'dist_center']]
y = data['price']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['age_owner', 'area', 'dist_center']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
# Создаем новый признак. Предположим, что есть некая зависимость цены от площади и защищённостью от зоиби
data['mult'] = (data['area'] * data['zombi_def'])
data.head(5)
```

```python
X = data[['mult']]
y = data['price']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['mult']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```
