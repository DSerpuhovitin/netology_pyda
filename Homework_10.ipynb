```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Russian Federation, Viet Nam, India, United States of America

# создаём дата фреймы и заменяенм NaN 
gii2017 = pd.read_csv('Analysis_2017.csv', delimiter=',').fillna('-')
gii2018 = pd.read_csv('Analysis_2018.csv', delimiter=',').fillna('-')

# переименовываем названия колонок, что бы избежать глюков
gii2017.rename(columns={' ': 'Index', ' .1': 'Subindex'}, inplace=True)
gii2018.rename(columns={' ': 'Index', ' .1': 'Subindex'}, inplace=True)

# создаю словарь, где ключи - страны, а значения - показатели 
data_2017 = pd.DataFrame({'Index' : list(gii2017['Index'][1:5].head()), 
            'Russian_Federation' : list(gii2017['Russian Federation'][1:5].head()),
            'Viet_Nam' : list(gii2017['Viet Nam'][1:5].head()),
            'India' : list(gii2017['India'][1:5].head()),
            'United_States_of_America' : list(gii2017['United States of America'][1:5].head())
})

data_2018 = pd.DataFrame({'Index' : list(gii2017['Index'][1:5].head()), 
            'Russian_Federation' : list(gii2018['Russian Federation'][1:5].head()),
            'Viet_Nam' : list(gii2018['Viet Nam'][1:5].head()),
            'India' : list(gii2018['India'][1:5].head()),
            'United_States_of_America' : list(gii2018['United States of America'][1:5].head())
})

```

```python
#строим графики

data_2017.hist()
data_2018.hist()
```
