```python
import pandas as pd
import numpy as np
```

```python
#к релизу переделать красиво
data2013 = pd.read_csv('short\\GII_2013.csv').drop(['rank','Income Group(Strength/Weakness)', 'Strength / Weakness', 'Value'], axis = 1)
data2013.rename(columns={'Score': 'Score_2013'}, inplace=True)

data2014 = pd.read_csv('short\\GII_2014.csv').drop(['rank','Income Group(Strength/Weakness)', 'Strength / Weakness', 'Value'], axis = 1)
data2014.rename(columns={'Score': 'Score_2014'}, inplace=True)

data2015 = pd.read_csv('short\\GII_2015.csv').drop(['rank','Income Group(Strength/Weakness)', 'Strength / Weakness', 'Value'], axis = 1)
data2015.rename(columns={'Score': 'Score_2015'}, inplace=True)

data2016 = pd.read_csv('short\\GII_2016.csv').drop(['rank','Income Group(Strength/Weakness)', 'Strength / Weakness', 'Value'], axis = 1)
data2016.rename(columns={'Score': 'Score_2016'}, inplace=True)

data2017 = pd.read_csv('short\\GII_2017.csv').drop(['rank','Income Group(Strength/Weakness)', 'Strength / Weakness', 'Value'], axis = 1)
data2017.rename(columns={'Score': 'Score_2017'}, inplace=True)

data2018 = pd.read_csv('short\\GII_2018.csv').drop(['rank','Income Group(Strength/Weakness)', 'Strength / Weakness', 'Value'], axis = 1)
data2018.rename(columns={'Score': 'Score_2018'}, inplace=True)

data2019 = pd.read_csv('short\\GII_2019.csv').drop(['rank','Income Group(Strength/Weakness)', 'Strength / Weakness', 'Value'], axis = 1)
data2019.rename(columns={'Score': 'Score_2013'}, inplace=True)
data2013.head()
```

```python
data = pd.merge(data2013, data2014, on='Economy')
data = data.merge(data2015, on='Economy')
data = data.merge(data2016, on='Economy')
data = data.merge(data2017, on='Economy')
data = data.merge(data2018, on='Economy')
data = data.merge(data2019, on='Economy')
data.rename(columns = {'Score_2013_x': 'Score_2013', 'Score_2013_y': 'Score_2019'}, inplace=True)
data.head()
```
